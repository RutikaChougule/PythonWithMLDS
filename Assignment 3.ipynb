```python
x = 2
lr = 0.01
precision = 0.000001
previous_step_size = 1
max_iter = 1000
iter = 0
gf = lambda x: (x + 3) ** 2
gd = []
```

```python
while precision < previous_step_size and iter < max_iter:
    prev = x
    x = x - lr*gf(prev)
    previous_step_size = abs(x - prev)
    iter += 1
    print("Iterations", iter, "Value", x)
    gd.append(x)
```

```python
print("Local minima for the given function is:", x)
```

```python
import matplotlib.pyplot as plt
plt.plot(gd)
```
